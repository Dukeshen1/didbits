# Gradient Descent in 1 dimension

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plot_poly(poly_func, min_x=-6.0, max_x=2.0):
    step = (max_x - min_x) / 200
    x = np.arange(min_x, max_x, step)
    y = poly_func(x)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x, y, 'r-')
    ax.set_xlabel('value of parameter')
    ax.set_ylabel('objective')
    return ax

def poly1(x):
    x2 = np.power(x, 2)
    x3 = np.power(x, 3)
    x4 = np.power(x, 4)
    y = x4 + 7*x3 + 5*x2 - 17*x + 3
    return y
```

```python
ax = plot_poly(poly1)
ax.text(-4, 40, '$x^4 + 7x^3 + 5x^2 - 17x + 3$')
plt.savefig('gradient-descent-1d.pdf')
```
